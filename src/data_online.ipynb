```python
import os
import pandas as pd
from tqdm import tqdm
```

```python
df_train_click = pd.read_csv('../dataset/train_click_log.csv')
df_test_click = pd.read_csv('../dataset/testA_click_log.csv')
df_train_click.describe(), df_test_click.describe()
```

```python
# 取出测试集中所有要计算的用户ID
test_users = df_test_click['user_id'].unique()
test_users
```

```python
test_query_list = []
for user in tqdm(test_users):
    test_query_list.append([user, -1])
test_query_list
```

```python
# 把上面的列表转化未DataFrame类型
df_test_query = pd.DataFrame(test_query_list, columns=['user_id', 'click_article_id'])
df_test_query
```

```python
# 把训练集和测试集的数据合在一起
df_query = df_test_query
df_click = pd.concat([df_train_click, df_test_click], sort=False).reset_index(drop=True)
df_train_click, df_test_click, df_click
```

```python
# 保存文件
os.makedirs('../data/online', exist_ok=True)

df_click.to_pickle('../data/online/click.pkl')
df_query.to_pickle('../data/online/query.pkl')
```
